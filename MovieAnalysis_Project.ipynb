Importing the required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib as mpl
import plotly.graph_objects as go
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
```

Importing the IMDB movies dataset

```python
dataset=pd.read_csv("IMDB Top 250 Movies.csv")
```

```python
genres_df = dataset['genre'].str.split(',', expand=True)
num_cols = len(genres_df.columns)
genres_df.columns = [f'genre_{i}' for i in range(1, num_cols+1)]
dataset = pd.concat([dataset, genres_df], axis=1)
```

```python
dataset
```

**Exploratory Data Analysis**

```python
df=dataset.copy()
```

Average Rating by Genre

```python
top_ratings=df.groupby('genre').mean().sort_values('rating',ascending=False).head(10)
top_ratings=top_ratings[['rating']].round(2)
top_ratings.reset_index(inplace=True)
top_ratings
```

Movie Ratings by Genre and Year

```python
grouped_df=df.groupby(['genre','year','rating']).size().reset_index(name='count')
figure=px.sunburst(grouped_df,path=['genre','year','rating'],values='count')
figure.update_layout(title='Movie Ratings by Genre and Year',height=1200,width=1200)
figure.show()
```

Director with most movies on the Top 250 list

```python
df_directors=df['directors'].str.split(',',expand=True).stack().reset_index(level=1,drop=True).rename('director')
df_top_directors=df_directors.value_counts().nlargest(10)
figure=px.bar(df_top_directors,x=df_top_directors.index,y=df_top_directors.values,labels={'x':'Director','y':'Number of Movies Directed'},color_discrete_sequence=['#8c6bb1'])
figure.update_traces(marker=dict(line=dict(color='black', width=1.25)))
figure.update_layout(title='Director vs Number of Movies Directed')
figure.show()
```

Wordcloud

```python
from wordcloud import STOPWORDS as stopwords_wc
from wordcloud import WordCloud,ImageColorGenerator
all_words=" ".join(token for token in dataset['genre'])
stopwords_wc=set(stopwords_wc)
wordcloud=WordCloud(stopwords=stopwords_wc,max_words=1500,max_font_size=350,random_state=42,width=2000,height=1000,colormap='twilight')
wordcloud.generate(all_words)
plt.figure(figsize=(16,8))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

Top 10 Directors and Their Best Movies

```python
df_directors=df.groupby('directors',as_index=False).agg({'rating':'mean'})
df_directors=df_directors.sort_values('rating',ascending=False).head(10)
df_best_movies=pd.DataFrame()
for director in df_directors['directors']:
  df_best_movies=df_best_movies.append(df[df['directors']==director].sort_values('rating',ascending=False).head(1))
  figure=go.Figure()
figure.add_trace(go.Bar(
  x=df_best_movies['name'],
  y=df_best_movies['rating'],
  text=df_best_movies['casts'],
  name='Best Movie',
  marker=dict(
      color='rgb(158,202,225)',
      line=dict(
          color='rgb(8,48,107)',
          width=1.5)
      ),
  opacity=0.6
))

figure.add_trace(go.Bar(
    x=df_directors['directors'],
    y=df_directors['rating'],
    name='Average Rating',
    marker=dict(
        color='rgb(255,142,104)',
        line=dict(
            color='rgb(8,48,107)',
            width=1.5)
        ),
    opacity=0.6
))

figure.update_layout(
    title='Top 10 Directors and Their Best Movies',
    xaxis_title='Movie',
    yaxis_title='Rating',
    barmode='group',
    bargap=0.15,
    bargroupgap=0.1,
    width=1000,
    height=500
)
```
